# Modelo de suavizado exponencial (ETS)

En primer lugar vamos a dividir la muestra en training y test. Dejamos fuera de la muestra los últimos 8 periodos, es decir, usaremos estos dos años como test. Empezamos construyendo la serie temporal:

```python
# Importamos las funciones necesarias:

from sktime.forecasting.base import ForecastingHorizon
from sktime.utils.plotting import plot_series
from sktime.forecasting.model_selection import temporal_train_test_split

# Dividimos la muestra
y_train, y_test = temporal_train_test_split(y = apple_ts['Ingresos'].astype('float64').to_period('Q'), test_size= 8)

# Dibujamos la partición:
plot_series(y_train, y_test, labels=["y_train", "y_test"])
print(y_train.shape[0], y_test.shape[0])
```

```python
# Pasamos a introducir el modelo ETS: En primer lugar cargamos:
from sktime.forecasting.ets import AutoETS
```

```python
# Creamos un horizonte de predicción con el tamaño del test (8)
fh = np.arange(1, 9)
apple_auto_model = AutoETS(auto= True, 
                           sp = 4, 
                           n_jobs=5)
```

```python
# Estimación del modelo, vamos a ver cuál elige el programa:
apple_auto_model.fit(y_train)
```

```python
# Le pedimos que nos muestre el modelo. Ha elegido MAM
print(apple_auto_model.summary())
```

Una vez elegido el modelo, queremos predecir los 8 periodos que hablábamos y además vamos a dibujarlo, para ver que preciso es. 

```python
# Predecimos: 
apple_pred = apple_auto_model.predict(fh)
print(apple_pred)
```

```python
# Dibujamos para ver como han ido las predicciones:
plot_series(y_train, apple_pred, y_test,
            labels= ["Apple", 'Apple pred', 'Apple real'],)
```

```python
plot_series(y_train['2015':], apple_pred, y_test,
            labels= ["Apple", 'Apple pred', 'Apple real'])
```

Efecto covid ? Podemos ver es positivo. La predicción ha sido bastante acertada para el periodo en el que tuvimos la declaración del estado de empergencia a nivel global, incluso llegando a ser, curiosamente, la predicción más acertada por parte de este modelo. 

```python
# Cálculo de las medias de error:
from sktime.performance_metrics.forecasting import mean_absolute_percentage_error
from sktime.performance_metrics.forecasting import MeanSquaredError

print("Mean absolute error is: ", mean_absolute_percentage_error(y_test, apple_pred))

```

```python
mse = MeanSquaredError()
mse(y_test, apple_pred)
```

```python
# Mean squared error:
rmse = MeanSquaredError(square_root= True)
rmse(y_test, apple_pred)
```

Podemos ver que aunque en los últimos dos periodos las predicciones son bastante imprecisas, durante el resto del tiempo el modelo se comporta bastante bien. Pasamos a predecir los periodos futuros a continuación:

```python
# Igual que antes, creamos el espacio temporal a predecir, 6 trimsestres:
# Jugamos con el mismo modelo que anteriormente.
# definimos y :

fh = np.arange(1, 7)
apple_auto_model = AutoETS(auto= True, sp= 4, n_jobs= -1)
y = apple_ts['Ingresos'].astype('float64').to_period('Q')
apple_auto_model.fit(y)
print(apple_auto_model.summary())
```

```python
# Visualizamos las predicciones, primero en tabla luego en gráfica:
apple_pred = apple_auto_model.predict(fh)
print(apple_pred)
```

```python
# Representación gráfica:
plot_series(y, apple_pred, labels= ["Apple: 1989-2021", "Predicción"])
```

Podemos ver que las predicciones basadas en este modelo son bastante optimistas, alcanzando a finales de este año 153 billones de ingresos y a finales del año que viene 194. 
Si comparamos estos datos con los de Barclays, difieren en: Q4 2021 ellos estiman 126 billones, y Q4 2022 134 billones. Sin embargo, es mucho más optimista para el primer cuatrimestre del año que viene, donde nuestro modelo arroja una cifra de 112 billones en comparación con los 191 de BArclays.

## Realizamos un suavizado exponencial 

```python
from sktime.forecasting.exp_smoothing import ExponentialSmoothing

forecaster = ExponentialSmoothing(trend='additive', seasonal='multiplicative', sp=4)
forecaster.fit(y)
```

```python
y_pred = forecaster.predict(fh)
y_pred
```

```python
print(forecaster._fitted_forecaster.summary())
```

```python
plot_series(y, y_pred, labels=["Apple", "Apple pred"])
```

```python
forecaster = ExponentialSmoothing(trend="add",seasonal="add",damped_trend=False, sp=4)

forecaster.fit(y)
y_pred = forecaster.predict(fh)

forecaster = ExponentialSmoothing(trend="add",seasonal="add",damped_trend=True, sp=4)
forecaster.fit(y)

y_pred_dump = forecaster.predict(fh)

plot_series(y["2015":], y_pred, y_pred_dump,labels=["Apple", "Apple pred","Apple Pred Dumpeded"])

```

Con el suavizado exponencial vemos que las predicciones parecen mucho más realistas que antes, usaremos estos datos para nuestro análisis. 

# PASAMOS A LA REALIZACIÓN DEL MODELO ARIMA

```python
from sktime.forecasting.base import ForecastingHorizon
from sktime.forecasting.model_selection import temporal_train_test_split
```

```python
apple_ts_q = apple_df['Ingresos'].astype('float64').to_period('Q').sort_index()
apple_ts_q.head()
```

```python
plot_series(apple_ts_q, labels=["Ventas apple"])
```

Transformación logarítmica de la serie original, para ver si la varianza se mantiene más o menos constante:

```python
from sktime.transformations.series.boxcox import LogTransformer
```

```python
transformer = LogTransformer()
log_apple_ts = transformer.fit_transform(apple_ts_q)
log_apple_ts.head()
```

```python
ax = log_apple_ts.plot(marker='o', linestyle='-')

ax.set_ylabel('Ventas Apple')
ax.set_title('Ventas Apple: Transformación LOG')
```

```python
# Lo mismo que arriba, pero con otro formato:
fig, ax = plot_series(log_apple_ts, labels=["Ventas"])
ax.set_title('Ventas Apple: Transformación LOG')
```

```python
# Podemos comparar este gráfico con las serie original:
fig, ax =plot_series(apple_ts_q, labels=["Ventas"])
ax.set_title('Ventas Apple: Serie Original')
fig, ax =plot_series(log_apple_ts, labels=["Ventas (log)"])
ax.set_title('Ventas Apple: Transformación LOG')
```

```python
# Pasamos a la realización del modelo:

y_train, y_test = temporal_train_test_split(y =apple_ts_q, test_size=8)

log_y_train, log_y_test = temporal_train_test_split(y =log_apple_ts, test_size=8)
plot_series(y_train, y_test, labels=["Train", "Test"])
```

```python
# Autocorrelación ?
plot_correlations(log_apple_ts)
```

```python
from sktime.transformations.series.difference import Differencer
transf_diff=Differencer(lags=[1])
plot_correlations(transf_diff.fit_transform(log_apple_ts))
```

```python
# Al igual que en el modelo de antes, fijamos el horizonte de predicción:
fh = np.arange(len(y_test)) + 1  
fh
```

```python
from sktime.forecasting.arima import AutoARIMA

forecaster = AutoARIMA(sp=4,suppress_warnings=True)
forecaster.fit(log_y_train)
```

```python
print(forecaster.summary())
```

```python
# Forecast realizado por ARIMA
log_y_pred = forecaster.predict(fh)
log_y_pred
```

```python
np.exp(log_y_pred)
```

```python
# Vamos a ver los errores de este modelo: MAPE
mean_absolute_percentage_error(log_y_test, log_y_pred)
```

```python
mean_absolute_percentage_error(y_test, np.exp(log_y_pred))
```

```python
rmse = MeanSquaredError(square_root=True)
rmse(log_y_test, log_y_pred)
```

```python
# Visualización de la predicción
plot_series(y_train, np.exp(log_y_pred),y_test, labels=["Apple", "Apple pred", "Apple REAL"])
```

```python
# Hacemos zoom:  
plot_series(y_train["2015":], np.exp(log_y_pred),y_test, labels=["Apple", "Apple pred", "Apple REAL"])
```

```python
# Predicción para datos futuros: 8 periodos
fh = np.arange(8) + 1  
fh
```

```python
forecaster = AutoARIMA(sp=4,suppress_warnings=True)
forecaster.fit(log_apple_ts)
```

```python
print(forecaster.summary())
```

```python
# Prediccion futura con el logaritmo
log_y_pred = forecaster.predict(fh)
log_y_pred
```

```python
# Predicción futura original:
np.exp(log_y_pred)
```

```python
plot_series(apple_ts_q, np.exp(log_y_pred), labels=["Apple", "Apple pred"])
```

```python
apple_yearly = apple_ts.resample('A').sum()
apple_yearly.tail( )
```

```python
# Nos sobra ya la columna de Q, que no aporta nada.
apple_yearly = apple_yearly.drop(['Q'], axis='columns')
```

```python
apple_yearly = apple_yearly.drop(['1989-12-31'])
apple_yearly.Crecimiento.dropna()
apple_yearly['Crecimiento'] = apple_yearly.Ingresos.pct_change().mul(100)
apple_yearly_mean = apple_yearly.mean()
print(apple_yearly.head())
print(apple_yearly.tail())
print(apple_yearly_mean)
```

```python
ax = apple_yearly['Crecimiento']['1991-12-31':'2021-06-30'].plot(marker='o', linestyle='-')
ax.set_title('Crecimiento de Apple en %')
ax.axhline(14.332, color='red', linestyle='dashed', linewidth = 1)
```
