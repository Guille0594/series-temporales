#### Quién sigue a quién ?


Intentando contestar a la pregunta que se nos realiza, analizamos los siguientes resultados. Hemos realizado un test de Granger, que nos dice si una serie temporal viene causada por otra. En la línea de código, la que introducimos primero es la que explica a la que introducimos en segundo lugar. Tenemos que fijarnos en el p valor, si nos encontráramos con un p valor muy bajo, podríamos decir que la primera serie explica, o provoca algún efecto en la segunda.

Primero comentamos a BMW, que parece no explicar a Volkswagen en ninguno de los períodos. Hemos puesto 13 lags para ver un año entero más un mes, ya que enero-enero(-1) podrían estar relacionados.

## 3. Modelo VAR
Realizamos una selección de VAR automática y vemos que nos arroja un modelo con dos ecuaciones. Hemos vuelto a usar el máximo de lags para que tenga todo ese período en cuenta.

Rápida explicación: ahora tenemos dos variables, nos habíamos centrado anteriormente en una. 
Tenemos, por ejemplo, manzanas y plátanos. Tenemos ventas de manzanas y de plátanos este mes y el mes del período anterior, t-1. El número de cada uno que vendí el período anterior, puede afectar a este. 

Lo interesante de estos modelos es que, también la venta de manzanas del mes anterior puede afectar a la venta de plátanos de este mes. También viceversa. 

#### Ecuación: at = C11 * At-1 + C12 * bt-1 + Ea1t
#### Siendo: at ventas de manzanas de este mes, c11 y c12 dos coeficientes, at-1 venta de manzanas del periodo anterior y bt-1 venta de plátanos del período anterior. E es el error en el número de manzanas que vendo este mes.

Por lo tanto vemos cómo la venta de manzanas de hoy está influenciada por la de ayer + la de plátanos de ayer. Para plátanos, sería similar. Ref: https://www.youtube.com/watch?v=UQQHSbeIaB0

Este sería un ejemplo de VAR 1 model porque sólo nos interesa un momento del pasado, si queremos más períodos las ecuaciones se complican más. Usamos vectores para estas fórmulas.


```python
# Para saber qúe lag debemos usar tenemos que mirar la partial autocorrelation 
# de nuestras series:
```

#### Correlación entre BMW y VK laggeado.

```python
# Creamos un buble para que, durante 13 meses, nos devuelva las correlaciones entre bmw y 
# volkswagen lageado todos esos periodos.  El primer vaor es la correlación, y el segundo
# es el p value. Como vemos, no hay ninguno significativo. Solamente quizás el 13, que sería
# el primer valor significante. Por eso usaremos 13 lags a continuación. 
for lag in range(1, 14):
    # definimos que es bmw y que es vk lageado:
    bmw_rend = stocks_ret['BMW.DE'].iloc[lag:]
    lagged_vk = stocks_ret['VOW.DE'].iloc[:-lag]
    # imprimimos coeficientes
    print('Lag: %s' %lag)
    print(pearsonr(bmw_rend, lagged_vk))
    print('---------')
```

#### Modelo VAR:

```python
# Creamos la variable modelo, que va a ser un VAR de stocks.
# Como hay dos columnas, las trata como time series
model = VAR(stocks_ret)
```

```python
# Va a usar un máximo de 13 lags para esta serie temporal.
model_fit = model.fit(13, ic='aic')
```

```python
# Como trata a las dos varaibles por igual, es por lo que nos devuelve resultados para 
# bmw y resultados para volkswagen. 
model_fit.summary()
```

 A continuación, estimamos nuestro modelo. Nos salen dos ecuaciones, un VAR(2). Nos fijamos en el prob, no en los coeficientes, que en estos modelos no nos importaba tanto. Hemos puesto 44 retardos para que nos coja el máximo posible.

## 4. Volatilidad de BMW
En este apartado pasamos a analizar la volatilidad de bmw y a la realización de un modelo GARCH. Cogemos para ello los datos diarios del mismo período y los representamos:

```python
# Para este estudio cogemos la data diaria de BMW para el mismo periodo:
daily_bmw = yf.download('BMW.DE',
                     start = start,
                     end = end, interval="1d")
daily_bmw = pd.DataFrame(daily_bmw['Adj Close'])
daily_bmw
```

```python
bmp_plot = daily_bmw.plot()
bmp_plot.set_ylabel('Adj Close');
```

A continuación, calculamos los rendimientos con la tasa de variación logarítmica. Los representamos. Después calcumaos los rendimientos al cuadrado, que son una aproximación a la volatilidad.

```python
# Vamos a ver los rendimientos 
daily_bmw['log_ret'] = np.log(daily_bmw['Adj Close']).diff()
daily_bmw.tail()
```

```python
# El siguiente paso es calcular los rendimientos al cuadrado
daily_bmw['rent_cuad'] = daily_bmw['log_ret']** 2
daily_bmw.tail()
```

Representación gráfica conjunta de precio, rentabilidad logarítmica y volatilidad

```python
# Vamos a visualizar la volatilidad. Vemos separados los gráficos de precio, rentablidad logarítmica 
# y volatilidad. Podemos ver que existe cierta clusterización en la volatilidad, lo cual es un indicador 
fig = plt.figure(figsize=(15,12))

plt.subplot(311)
ax = daily_bmw['Adj Close'].plot( linestyle='-')
ax.set_ylabel('Precio')
ax.set_xlabel('')
ax.set_title('BMW', y=1, fontsize = 16)

plt.subplot(312)
ax = daily_bmw['log_ret'].plot( linestyle='-')
ax.set_ylabel('Rentabilidad')
ax.set_xlabel('')
ax.set_title('Rentabilidad', y=0.8, x=0.6,fontsize = 16)

plt.subplot(313)
ax = daily_bmw['rent_cuad'].plot( linestyle='-')
ax.set_ylabel('BMW')
ax.set_xlabel('')
ax.set_title('Volatilidad', y=0.8, fontsize = 16)
```

Podemos observar ciertos clústeres en la volatilidad de BMW, lo cual no es un buen indicador. Hacemos más pruebas. Viéndola gráficamente es un buen candidato para el proceso Garch. Lo importante es saber que orden de Garch darle. Para ello,vamos a fijarnos en la autocorrelación de la volatilidad: PACF returns ** 2. Lo hacemos dos celdas más abajo.

```python
# Pasamos a calcular la volatilidad de bmw: 
bmw_vol = pd.DataFrame(daily_bmw['rent_cuad'].astype('float64').to_period('D').sort_index().dropna())
bmw_vol.tail()
```

### ¿Hay autoccorelación?

```python
# Podemos comprobar que sí existe correlación en la volatilidad de BMW, por lo tanto la z al cuadrado 
# sigue teniendo comportamiento Garch, así que ste modelo no es adecuado para predecir. 
plot_correlations(bmw_vol)
```

Podemos ver que existe mucha autocorrelación en la volatilidad, pero de momento cogeremos un Garch 2, 2 puesto que a partir del segundo de alguna forma empieza a caer. Hay más momento, en los que existe autocorrelación, pero para mantenerlo siemple cogemos esos 2. Veremos que al final para este trabajo nos decantamos por un Garch 1, 1. 

```python
# Importamos la función de test Ljung_Box
import statsmodels.api as sm

Ljung_Box = sm.stats.acorr_ljungbox 
```

```python
# La hipótesis nula es que NO existe autocorrelación. Como tenemos pvalores muy bajos, podemos descartar
# esa hipótesis nula y decir que si que existe autocorrelación. Aunque ya lo vimos biengráficamente. 
Ljung_Box(bmw_vol, lags = 15, return_df=True)
```

```python
# Probamos con otro test el mismo concepto:
import statsmodels.api as sm
test_LM_ARCH = sm.stats.diagnostic.het_arch
```

```python
# Igualmente tenemos un p muy bajo, sí hay autocorrelción. 
lm, lmpval, f, pval, result_int = test_LM_ARCH(daily_bmw['log_ret'].dropna(), nlags = 13, store=True)
print ('Estadístico LM: ', lm)
print ('p-value: ', lmpval)
print ('Estadístico F: ', f)
print ('p-value: ', pval)
```

### GARCH (1, 1)

Empezamos modelando con un Garch (1,1). Antes hemos probado con un GARCH 2,2 como hemos señalado antes. El problema es que el p valor de la segunda alfa y la segunda beta no era significativo, por eso finalmente nos quedarmos con un 1, 1, puesto que sumando alfa y beta nos da un valor por debajo de 1. 

```python
# Estimamos nuestro modelo Garch:
ar = ARX(100*daily_bmw['log_ret'].dropna(), lags = 350)
print(ar.fit().summary())
```

```python
# recordemos que se tenía que cumplir que alfa1 y beta1 suamdos fueran menores a 1, y en este caso estamos al 
# límite, pues el resultado es 0.99. Alfa es el lag de la time series y omega de la volatilidad.
ar.volatility = GARCH(p=1 , q=1) # p siendo alfa, q omega
res = ar.fit(update_freq=0, disp="off")
print(res.summary())
```

```python
# Comprobamos ahora, si se siguen comprobando aquellos clústeres en la volatilidad: 
# Comprobamos que se nota menos que antes, pero tenemos que verlo numéricamente ahora. 
fig = res.plot(annualize="D")
```

```python
# RESIDUOS:
bmw_std_resid = (res.resid / res.conditional_volatility).astype('float64').to_period('D').sort_index().dropna()
```

Podemos observar en la siguiente gráfica que la autocorrelación en los residuios de la volatilidad ha caído, ya no hay. Esto es un buen idnicador, lo siguiente que debemos comprobar es cómo se ajusta la z al cuadrado a la diagonal. 

```python
plot_correlations(bmw_std_resid)
```

```python
plot_correlations(bmw_std_resid**2)
```

```python
# Veamos la voaltilidad.
fig = res.plot(annualize="D")
```

En los extremos se separa un poco, por lo que a continuación probaremos con la T Student.

```python
# El modelo no es correcto, y como tal es probable que no lleve una normal. Vamos a verlo:
# Se ajusta bien, pero vamos a verlo con la T student a continuación.
sm.qqplot(bmw_std_resid,line='45')
plt.show()
```

### T Student

```python
from arch.univariate import StudentsT
```

```python
ar2 = ARX(100*daily_bmw['log_ret'].dropna(), lags= 13)
print(ar2.fit().summary())
```

```python
ar2.distribution = StudentsT()
res2 = ar2.fit(update_freq=0, disp="off")
print(res2.summary())
```

```python
# lo mismo que antes
fig = res2.plot(annualize="D")
```

```python
# Residuos
bmw2_std_resid=(res2.resid / res2.conditional_volatility).astype('float64').to_period('D').sort_index().dropna()
```

```python
# Plot Autocorrelation
plot_correlations(bmw2_std_resid)
```

```python
# Plot Autocorrelation. vemos que sigue sin haber efecto garch
plot_correlations(bmw2_std_resid**2)
```

Aunque exista más autocorrelación, se ajusta perfectamente a la recta, y este criterio tiene más peso que la autocorrelación. Por lo tanto nos quedamos con este modelo para las predicciones de noviembre que hacemos a continuación.

```python
# Con la T de Student nos queda algo así, muchísimo mejor que antes.
from scipy import stats
sm.qqplot(bmw2_std_resid, dist = stats.t,fit = True, line='45')
plt.show()
```
